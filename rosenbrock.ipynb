```python
import numpy as np
from scipy.optimize import minimize

# Rosenbrock function definition
def rosenbrock(x):
    a = 1
    b = 100
    return (a - x[0])**2 + b * (x[1] - x[0]**2)**2

# Initial guess
x0 = np.array([0.0, 0.0])
```

```python
# Minimize the Rosenbrock function
result = minimize(rosenbrock, x0, method='BFGS')
```

```python

# Optimal solution
x_opt, y_opt = result.x
obj_opt = result.fun

print(f"Optimal solution: x = {x_opt}, y = {y_opt}")
print(f"Objective value at optimal solution: {obj_opt}")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
a = 1
b = 100

# Create a grid of points
X = np.linspace(-2, 2, 400)
Y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(X, Y)
Z = (a - X)**2 + b*(Y - X**2)**2
```

```python

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')


# Plot the optimal point
ax.scatter(x_opt, y_opt, obj_opt, color='r', s=50, label='Optimal Point')
ax.legend()
plt.show()
```
